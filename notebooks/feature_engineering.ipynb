```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

user_feats = pd.read_csv("../intermediate/user_features.csv")
user_feats.head()
```

```python
numeric_cols = ["total_transactions", "avg_amount", "monthly_freq", "tx_count_30d",
                "pct_crypto", "pct_stocks", "pct_etf", "pct_bond", "pct_forex", "pct_other",
                "volatility_mean_30", "volatility_mean_60", "volatility_std_30"]

for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(user_feats[col], kde=True, bins=30)
    plt.title(f"{col} Distribution")
    plt.show()
```

Histogram Analysis â€“ Numeric Features

1. total_transactions

Most users have a low number of transactions, while a few users have very high counts.

The distribution is rightskewed, indicating that a small segment of highly active users may need separate consideration in analysis or segmentation.

2. avg_amount

The average transaction amounts are spread over a wide range.

Smaller to medium sized transactions are common, whereas very large transactions are rare.

Outliers in transaction amounts could influence modeling and should be considered.

3. monthly_freq

Monthly transaction frequency is low for the majority of users.

Only a subset of users is actively trading each month.

Useful feature for capturing user engagement intensity.

4. tx_count_30d

Most users have zero or very few transactions in the last 30 days, highlighting recent activity patterns.

This feature is critical for short-term behavioral modeling.

5. pct_crypto, pct_stocks, pct_etf, pct_bond, pct_forex, pct_other

Usersâ€™ portfolio allocations are heterogeneous.

Stocks and crypto dominate user holdings, while ETFs, bonds, and forex have lower representation.

Skewed allocation can indicate varying risk preferences among users.

6. volatility_mean_30, volatility_mean_60, volatility_std_30

Average and standard volatility values show considerable variation across users.

Users with higher risk profiles tend to invest in assets with higher volatility.

These features are valuable signals for risk modeling.

```python
risk_levels = ['low', 'medium', 'high']

for col in["avg_amount", "monthly_freq", "pct_crypto", "pct_stocks", "volatility_mean_30"]:
    plt.figure(figsize=(6,4))
    sns.violinplot(x="risk_level", y=col, data=user_feats, order=risk_levels)
    plt.title(f"{col} by Risk Level")
    plt.show()
```

Violin Plot Analysis â€“ Features by Risk Level

1. avg_amount by Risk Level

High-risk users generally have higher average transaction amounts compared to medium and low risk users.

Low risk users tend to transact smaller amounts.

Suggests a positive relationship between transaction size and risk appetite.

2. monthly_freq by Risk Level

High risk users tend to transact more frequently each month than medium and low-risk users.

Low risk users show lower monthly transaction frequency.

Monthly activity is a good indicator of risk engagement behavior.

3. pct_crypto by Risk Level

High risk users allocate a higher proportion of their portfolio to cryptocurrencies.

Low risk users have minimal crypto exposure.

Confirms that crypto holdings correlate strongly with higher risk tolerance.

4. pct_stocks by Risk Level

Low risk users tend to have a higher percentage in stocks relative to high-risk users.

Medium and high risk users diversify more into crypto and other volatile assets.

Indicates traditional equities are preferred by conservative investors.

5. volatility_mean_30 by Risk Level

High risk users are exposed to higher 30-day average volatility in their portfolio.

Low risk users experience much lower portfolio volatility.

This aligns with the expected behavior: risk tolerant users engage with more volatile assets.

```python
plt.figure(figsize=(12,10))
sns.heatmap(user_feats[numeric_cols].corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Feature Correlation Matrix")
plt.show()
```

Feature Correlation Matrix Analysis

total_transactions & monthly_freq / tx_count_30d

Strong positive correlation, as expected: users who transact more overall also transact more frequently both monthly and in the last 30 days.

Confirms that these features capture overlapping aspects of user activity.

avg_amount

Weak correlation with transaction counts, suggesting that high frequency traders donâ€™t necessarily make large transactions.

Highlights the independence of transaction size from transaction frequency.

Asset allocation features (pct_crypto, pct_stocks, pct_etf, pct_bond, pct_forex, pct_other)

Mostly negatively correlated with each other (except for small positive correlations in some cases), because allocations sum to 1.

Crypto allocation has mild negative correlation with stocks, indicating risk-seeking users prefer crypto over traditional equities.

Volatility features (volatility_mean_30, volatility_mean_60, volatility_std_30)

High positive correlation among volatility metrics, as expected.

Users exposed to volatile assets in 30 days are also likely exposed to volatility in 60 days.

Weak correlations with transaction counts, showing that risk taking in assets is somewhat independent of activity levels.

Overall Insights:

Activity based features cluster together (frequency, count, recency).

Asset allocation features are inversely related, reflecting portfolio composition trade offs.

Volatility measures are consistent and can serve as strong signals in risk modeling.

```python
avg_pct_cols = ["pct_crypto", "pct_stocks", "pct_etf", "pct_bond", "pct_forex", "pct_other"]
avg_pct = user_feats[avg_pct_cols].mean().sort_values(ascending=False)

plt.figure(figsize=(6,6))
avg_pct.plot(kind="pie", autopct='%1.1f%%', startangle=90)
plt.title("Average Asset Allocation Across Users")
plt.ylabel("")
plt.show()
```

Average Asset Allocation Across Users

Portfolio Composition

The chart shows the average proportion of each asset type across all users.

Stocks and crypto dominate the portfolios, indicating that users favor these assets.

ETFs, bonds, forex, and other assets have smaller shares on average.

Insights

Users are generally risk seeking, as crypto holds a significant portion.

Low allocation to bonds and other conservative instruments suggests the user base is skewed towards growth and high return assets.

Portfolio diversification is moderate: while some variety exists, a few asset types dominate the overall allocation.

Implications for Risk Modeling

High exposure to volatile assets like crypto can be strong predictors of higher risk levels.

Stock heavy portfolios may correlate with medium risk users, while low exposure to traditional bonds indicates fewer conservative users.
